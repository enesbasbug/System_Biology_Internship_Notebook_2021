## Benjamini & Hochberg (False discovery rate) and cut-off = 0.05

```python
import pandas as pd
import numpy as np

data = pd.read_csv("GSE175361.top.table.tsv", sep='\t')
data.head()
```

```python
data = data.drop(['RANGE_START','RANGE_END','SPOT_ID','RANGE_GB'], axis = 1)
data
```

if : p-value >= cut-off 

--> accept 'No significant changes in the expression levels' [ H0 ]
    
    OR
    
--> accept 'Potential significant changes (upregulation/downregulation)' [ H1 ]



```python
data["Any_Changes"] = data["adj.P.Val"].apply(lambda x: 'H0 : No significant changes' if x >= 0.05 else 'H1 : Potential significant changes ')
data[10:20]
```
