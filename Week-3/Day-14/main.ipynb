```python
import pandas as pd

data = pd.read_csv("GSE31432.top.table.tsv", sep='\t')
data.head()
```

```python
data.describe()
```

```python
data = data.drop(axis=1, columns = ["t","B"])
```

```python
data['Gene.symbol'].isnull().values.sum()
```

```python
data.dropna(subset = ["Gene.symbol"], inplace=True)
data.head()
```

```python
data['Gene.symbol'].isnull().values.sum()
```

```python
data_2 = data.drop(axis=1, columns=["ID","P.Value","logFC","Gene.title"])
data_2.head()
```

```python
cols = list(data_2.columns)
cols = cols[1::2] + cols[::2]
data_2 = data_2[cols]
```

```python
data_2.head()
```

```python
data_3 = data_2[data_2['Gene.symbol'].str.contains("/")] # we dont want these '///' names in our dataset
data_3
```

```python
delete = data_3.index.tolist()
delete
```

```python
data_2 = data_2.drop(delete, axis=0)
```

```python
data_2[data_2['Gene.symbol'].str.contains("/")].sum() # we dont have long gene.symbol names with '///'. I've deleted them all
```

```python
data_2['Gene.symbol'].isnull().values.sum()
```

```python
data_2.head()
```

```python
data_2.to_csv('data_symbol_pval.txt', index=False)
```

## Gene ID and Binary Table

```python
data_4 = pd.read_csv("bioDBnet_db2db_210812143008_453561945.txt", sep='\t')
data_4
```

```python
frames = [data_3, data_4]
  
data_5 = pd.concat(frames, axis=1, join='inner')
data_5.head()
```

```python
data_5.columns = [c.replace(' ', '_') for c in data_5.columns]
data_5.head()
```

```python
data_5.drop(['Gene.symbol','Gene_Symbol', 'Unnamed:_2'], axis=1, inplace=True)
data_5.head()
```

```python
data_5.to_csv('expression_data_geneId_binary.txt', index=False, sep="\t")
```
